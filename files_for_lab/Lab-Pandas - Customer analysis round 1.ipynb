```python
file3 = pd.read_csv('file3.csv')
file3
```

```python
column_names = file1.columns
```

```python
#data set created, blank
data = pd.DataFrame(columns=column_names)
data.head()
```

```python
data.shape
```

```python
#rearrange columns for file2
file2 = file2[['Customer','ST','GENDER','Education','Customer Lifetime Value', 'Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Vehicle Class','Total Claim Amount']]
```

```python
#file3, rename ST and Gender > ST and GENDER to fit data set (standardize names later)
file3.head()
```

```python
file3 = file3.rename(columns={'State':'ST','Gender':'GENDER'})
```

```python
file3 = file3[['Customer','ST','GENDER','Education','Customer Lifetime Value', 'Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Vehicle Class','Total Claim Amount']]
```

```python
#stitching data together / blend data
```

```python
data = pd.concat([data,file1,file2,file3], axis=0)
```

```python
data = data.rename(columns={'ST':'state','GENDER':'gender'})
```

```python
data
```

```python
# Standardizing header names
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower())
    
data.columns = cols
```

```python
#show dataframe shape
data.shape
```

```python
data.dtypes
```

```python
#delete columns: ediucation & number of open complaints
data = data.drop(['education','number of open complaints'],axis=1)
```

```python
#create copy before data cleaning process
temp = data.copy() 
```

```python
#attempting to remove the % from customer lifetime value, if it exists > changes to numerical
data['customer lifetime value'] = pd.to_numeric(data['customer lifetime value'].astype(str).str[:-1], errors='coerce')
```

```python
#correct values for customer lifetime value: multiply by 100
data['customer lifetime value'] = data['customer lifetime value'].apply(lambda x:x*100)
```

```python
print(data.shape)
data = data.drop_duplicates()
print(data.shape)
```
